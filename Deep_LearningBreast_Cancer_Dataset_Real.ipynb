## **Standerdize The Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

# **Let's Build A Neural Network from the data**

```python
# Importing the the neccesery liberies for a neural network

import tensorflow as  tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
from keras.api._v2.keras import activations
# now lets set up the layers for the neural network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')

])
```

```python
# compiling the neural network

model.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

# **Let's train the neural network**

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=15)
```

# **Visualizing accuracy and loss**

```python
# let's see the model loss in a line graph

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['training data', 'validation data'], loc = 'lower right')
plt.show()
```

```python
# let's see the model accuracy in a line graph


plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training data', 'validation data'], loc = 'lower right')
plt.show()
```

# **Accuracy of the model on the test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
# The Input data below is from the dataset

input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change in input data to a numpy array

input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we predicting for ine data point

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0]==0):
  print('The Tumor is Malignant')

else:
  print('The Tumor is Benign')
```
